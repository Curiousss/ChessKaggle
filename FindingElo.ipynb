```python
from sklearn.feature_extraction import DictVectorizer
from collections import defaultdict, Counter
import chess.pgn
import chess
from chess import QUEEN, KING
import numpy as np
import pandas as pd
```

```python
def extract_game_features(game):
    """Extract features for a given game"""

    first_check = True
    first_queen_move = True
    first_king_move = True

    features = defaultdict(int)
    node = game

    while node.variations:
        
        move = node.variation(0).move

        board = node.board()
        
        if(board.turn == chess.WHITE):
            turn = "W_"
        else:
            turn = "B_"

        moved_piece = board.piece_type_at(move.from_square)
        captured_piece = board.piece_type_at(move.to_square)

        if moved_piece == QUEEN and first_queen_move:
            features[turn+'queen_moved_at'] = board.fullmove_number
            first_queen_move = False

        if captured_piece == QUEEN:
            features[turn+'queen_changed_at'] = board.fullmove_number

        # DOES THE NUMBER OF TIMES THE KING IS MOVED MATTER OTHER THAN FOR CASTLING?
        if moved_piece == KING and not board.is_castling(move):
            if first_king_move:
                features[turn+'king_moved_at'] = board.fullmove_number
                first_king_move = False
            features[turn+'no_king_moves'] += 1

        if move.promotion:
            features[turn+'promotion'] += 1
        if board.is_check():
            features[turn+'total_checks'] += 1
            if first_check:
                features[turn+'first_check_at'] = board.fullmove_number
                first_check = False
        # castling, can be done only once
        if(board.is_kingside_castling(move) == True):
            features[turn+"king_side_caslte"] = board.fullmove_number
        elif(board.is_queenside_castling(move) == True):
            features[turn+"is_queenside_castling"] = board.fullmove_number
        
        if board.can_claim_draw():
            features[turn+'can_claim_draw'] += 1
        if board.can_claim_fifty_moves():
            features[turn+'fifty_moves'] += 1
        if board.can_claim_threefold_repetition():
            features[turn+'threefold_rep'] += 1
        if board.is_fivefold_repetition():
            features[turn+'fivefold_rep'] += 1
        if board.is_seventyfive_moves():
            features[turn+'seventyfive_moves'] += 1
        
        #NEXT MOVE
        node = node.variation(0)
        
    # END OF THE GAME
    # Donno if it helps capturing these features specific to White/Black turn
    if board.is_checkmate():
        features[turn+'is_checkmate'] += 1
    if board.is_stalemate():
        features[turn+'is_stalemate'] += 1
    if board.is_insufficient_material():
        features[turn+'insufficient_material'] += 1
    if board.is_game_over():
        features[turn+'is_game_over'] += 1
    features['total_moves'] = board.fullmove_number

    # Pieces at the end of the game
    piece_placement = board.fen().split()[0]
    end_pieces = Counter(x for x in piece_placement if x.isalpha())

    # count number of piece at end position
    features.update({'end_' + piece: cnt
                     for piece, cnt in end_pieces.items()})
    return features
```

```python
def games_features():
    l = []
    pgn = open("../givenData/data/data.pgn")
    #game = chess.pgn.read_game(pgn)
    cnt = 0
    elos = []
    results = []
    resultFeature = {}
    while cnt < 50000:
        print(cnt)
        game = chess.pgn.read_game(pgn)
        features = extract_game_features(game)
        result = game.headers['Result'].split("-")
        if (result[0] == '1/2'):
            result[0] = 0.5
            result[1] = 0.5
        else:
             result[0] = int(result[0])
             result[1] = int(result[1])
        features['WhiteResult'] = result[0]
        features['BlackResult']= result[1]
        results.append(resultFeature)
        l.append(features)
        if cnt < 25000:
            elos.append(int(game.headers['WhiteElo']))
            elos.append(int(game.headers['BlackElo']))
        cnt += 1
    return l, elos, results
l, elos, results = games_features()
```

```python
vec = DictVectorizer()
X = vec.fit_transform(l)

names=vec.get_feature_names()
games_array = pd.DataFrame(X.toarray(), columns=names)
```

```python
def extract_score_features(scores):
    features = dict()
    if scores.size == 0:
        return features
    scores = np.r_[0, scores]
    abs_scores = np.abs(scores)
    diffs = np.diff(scores)
    white_diffs = diffs[::2]
    black_diffs = diffs[1::2]
    subset_names = ['diffs', 'white_diffs', 'black_diffs']
    subsets = [diffs, white_diffs, black_diffs]
    stats = [np.min, np.max, np.mean, lambda x: np.median(np.abs(x))]
    stat_names = ['min', 'max', 'mean', 'median_abs']
    for subset, subset_name in zip(subsets, subset_names):
        for stat, stat_name in zip(stats, stat_names):
            features[stat_name + '_' + subset_name] = stat(subset)
            # np.hi
    features['advantage120_idx'] = np.argmax(abs_scores > 120) or len(scores)
    features['advantage70_idx'] = np.argmax(abs_scores > 70) or len(scores)
    return features
```

```python
l = stockfish.MoveScores.apply(extract_score_features).tolist()
vec = DictVectorizer()
X = vec.fit_transform(l)
scores_df = pd.DataFrame(X.toarray(), columns=vec.get_feature_names())
```

```python
def baseline_model():
 # create model
    model = Sequential()
    model.add(Dense(64, input_dim=40, kernel_initializer='normal', activation='relu'))
    model.add(Dense(32, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal'))
    # Compile model
    model.compile(loss='mean_absolute_error', optimizer='adam')
    return model
```
